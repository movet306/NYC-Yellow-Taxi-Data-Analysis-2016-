```python
import pandas as pd

df_daily_trips = pd.read_csv("daily_trip_count_nyc_2016.csv.csv")
df_hourly_trips = pd.read_csv("hourly_trip_count.csv`.csv")
df_hourly_tip = pd.read_csv("hourly_tip_rate.csv.csv")
df_tip_by_payment = pd.read_csv("tip_rate_by_payment_type2.csv")
df_tip_by_distance = pd.read_csv("Tip Rate by Trip Distance.csv")
df_trip_duration = pd.read_csv("average trip duration.csv")
df_trips_by_weekday = pd.read_csv("Number of Trips by Day of the Week.csv")
df_trips_by_month = pd.read_csv("Number of Trips by Month.csv")
df_hourly_trip_count = pd.read_csv("hourly_trip_count.csv")
df_day_hour = pd.read_csv("Number of Trips by Day of Week and Hour.csv")
df_trip_tip_duration = pd.read_csv("trip_tip_duration.csv.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_daily_trips['trip_date'] = pd.to_datetime(df_daily_trips['trip_date'])

plt.figure(figsize=(14, 6))
sns.set_style("whitegrid")

sns.lineplot(data=df_daily_trips, x='trip_date', y='trip_count', color='royalblue')

plt.title("🗓️ Daily Yellow Taxi Trips in NYC (2016)", fontsize=16, fontweight='bold')
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Trips", fontsize=12)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grafik ayarı
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Saatlik grafik
sns.barplot(data=df_hourly_trip_count, x='pickup_hour', y='trip_count', color='royalblue')

# Başlık ve eksen etiketleri
plt.title("⏰ Hourly NYC Yellow Taxi Trips in 2016", fontsize=16, fontweight='bold')
plt.xlabel("Hour of Day (0–23)", fontsize=12)
plt.ylabel("Number of Trips", fontsize=12)

# X-ticks düzenle
plt.xticks(ticks=range(0, 24), labels=[str(i) for i in range(0, 24)])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

sns.barplot(data=df_hourly_tip, x='pickup_hour', y='avg_tip_rate', color='mediumseagreen')

plt.title("💸 Average Tip Rate by Hour of Day", fontsize=16, fontweight='bold')
plt.xlabel("Pickup Hour (0–23)", fontsize=12)
plt.ylabel("Average Tip Rate", fontsize=12)

plt.xticks(ticks=range(0, 24), labels=[str(i) for i in range(0, 24)])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

sns.barplot(data=df_tip_by_payment.sort_values(by='avg_tip_rate', ascending=False),
            x='avg_tip_rate', y='payment_method', palette='viridis')

plt.title("💳 Average Tip Rate by Payment Method", fontsize=16, fontweight='bold')
plt.xlabel("Average Tip Rate", fontsize=12)
plt.ylabel("Payment Method", fontsize=12)
plt.xlim(0, 0.25)

for index, row in df_tip_by_payment.iterrows():
    plt.text(row['avg_tip_rate'] + 0.005, index, f"{row['avg_tip_rate']:.2%}", va='center')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

bar = sns.barplot(data=df_tip_by_distance, 
                  x='distance_bucket', 
                  y='avg_tip_rate', 
                  color='cornflowerblue')

plt.title("📏 Average Tip Rate by Trip Distance", fontsize=16, fontweight='bold')
plt.xlabel("Trip Distance", fontsize=12)
plt.ylabel("Average Tip Rate", fontsize=12)

for i in bar.containers[0]:
    value = i.get_height()
    bar.text(i.get_x() + i.get_width()/2, value + 0.002, 
             f"{value*100:.2f}%", 
             ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 5))
sns.set_style("whitegrid")

sns.barplot(data=df_trip_duration, x=['Average Trip Duration'], y='avg_trip_duration_min', color='mediumseagreen')

# Annotate value on top
duration = df_trip_duration['avg_trip_duration_min'].values[0]
plt.text(0, duration + 0.5, f"{duration:.1f} mins", ha='center', fontsize=12, fontweight='bold')

plt.title("⏱️ Average Yellow Taxi Trip Duration (NYC - 2016)", fontsize=14, fontweight='bold')
plt.ylabel("Minutes")
plt.xticks([])  

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

weekday_map = {1: 'Sun', 2: 'Mon', 3: 'Tue', 4: 'Wed', 5: 'Thu', 6: 'Fri', 7: 'Sat'}
df_trips_by_day['day_name'] = df_trips_by_day['day_of_week'].map(weekday_map)

order = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']

bar = sns.barplot(data=df_trips_by_day, x='day_name', y='trip_count', order=order, color='goldenrod')

plt.title(" Number of Trips by Day of the Week", fontsize=16, fontweight='bold')
plt.xlabel("Day of Week", fontsize=12)
plt.ylabel("Number of Trips", fontsize=12)

for i in bar.containers[0]:
    value = i.get_height()
    bar.text(i.get_x() + i.get_width()/2, value + 10000, f"{int(value):,}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort month column if needed
df_trips_by_month = df_trips_by_month.sort_values(by='month')

month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

bar = sns.barplot(
    data=df_trips_by_month,
    x='month',
    y='trip_count',
    palette='Blues_d'
)

for p in bar.patches:
    height = p.get_height()
    bar.annotate(f"{int(height):,}", 
                 (p.get_x() + p.get_width() / 2., height), 
                 ha='center', va='bottom', 
                 fontsize=9, color='black', 
                 xytext=(0, 5), textcoords='offset points')

# Set titles and labels
plt.title("📅 Monthly Yellow Taxi Trips in NYC (2016)", fontsize=16, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Trips", fontsize=12)
plt.xticks(ticks=range(12), labels=month_labels)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_day_hour = pd.read_csv("Number of Trips by Day of Week and Hour.csv")

heatmap_data = df_day_hour.pivot(index="pickup_hour", columns="day_of_week", values="trip_count")

plt.figure(figsize=(12, 7))
sns.set_style("whitegrid")
sns.heatmap(
    heatmap_data,
    cmap="YlOrRd",
    linewidths=0.5,
    annot=False,
    fmt="g",
    cbar_kws={"label": "Number of Trips"}
)

plt.title("🧭 NYC Yellow Taxi Trip Density by Hour & Day (2016)", fontsize=16, fontweight='bold')
plt.xlabel("Day of Week", fontsize=12)
plt.ylabel("Hour of Day", fontsize=12)
plt.yticks(rotation=0)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
